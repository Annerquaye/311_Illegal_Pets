```python
# #Now we can open our data which downloaded from data.nyc 
# using descriptor == "Illegal Animals kept as pets". We chose from 2010 to year-to-date.

df = pd.read_csv("illegal_animals_kept_as_pets1.csv")
```

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
illegal_pets = df
```

```python
#Next we can check to see which types of animals are often reported via the 'descriptor'.


illegal_pets.descriptor.value_counts()
```

```python
#We convert the dates into datetime in order to be able to use it to 
#plot if need be.

illegal_pets['closed_date'] = pd.to_datetime(illegal_pets.closed_date, format="%m/%d/%Y %I:%M:%S %p", errors='coerce')
illegal_pets['created_date'] = pd.to_datetime(illegal_pets.created_date, format="%m/%d/%Y %I:%M:%S %p", errors='coerce')
illegal_pets.head()
```

```python
illegal_pets.descriptor.value_counts()
```

```python
# Now we can check which boroughs has most complaints in terms of all animals
# or in terms of specific animals

#To check for total complaints by borough
illegal_pets.borough.value_counts() 

#To check which animals is dominates by borough
illegal_pets.groupby(by='descriptor').borough.value_counts() 
```

```python
#This is to plot the weekly reports of complaints

illegal_pets.resample('W', on='created_date').size() \
    .plot(title="311 complaints of Illegal Animals kept as pets since 2010")
```
